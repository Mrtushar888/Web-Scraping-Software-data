```python
import numpy as np
import requests
import pandas as pd
from bs4 import BeautifulSoup

webpage=requests.get('https://filehippo.com/popular/1/').text

soup=BeautifulSoup(webpage,'lxml')
software=  soup.find_all('a',class_='card-program')
names = []
License = []
About = []
Links = []
company = []

for i in software:
    names.append(i.find('p', class_='card-program__title').text)
    company.append(i.find('span', class_='card-program__developer').text)
    License.append(i.find('span', class_='card-program__license').text)
    About.append(i.find('p', class_='card-program__summary').text)
    href=i['href']
    if href not in Links:
        Links.append(href)

print(company)

naming={"Software":names,"Company":company,"License":License,"About":About,"Link":Links}
df=pd.DataFrame(naming)
df

Final_scrap=[]
for j in range(1,998):
    webpage=requests.get(f'https://filehippo.com/popular/{j}/').text
    soup=BeautifulSoup(webpage,'lxml')
    software=  soup.find_all('a',class_='card-program')
    names = []
    License = []
    About = []
    Links = []
    company = []

    for i in software:
        names.append(i.find('p', class_='card-program__title').text)
        company_elem = i.find('span', class_='card-program__developer')
        company.append(company_elem.text if company_elem else "")
        License.append(i.find('span', class_='card-program__license').text)
        About.append(i.find('p', class_='card-program__summary').text)
        href=i['href']
        if href not in Links:
            Links.append(href)
    naming={"Software":names,"Company":company,"License":License,"About":About,"Link":Links}
    df=pd.DataFrame(naming)
    Final_scrap.append(df)
final_df = pd.concat(Final_scrap, ignore_index=True)
final_df=final_df.drop_duplicates()
final_df = final_df.reset_index(drop=True)
print(final_df)

final_df.describe
final_df.tail(10)
final_df.to_csv('filehippo.csv',index=False)
```
